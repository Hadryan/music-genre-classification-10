```python
import pandas as pd
import utils
import os
```

```python
root_path = '/Users/saurabh/PycharmProjects/IDS/final_project/music-genre-classification/data/2018655052_small'
tracks_dict = {}
for path, subdirs, files in os.walk(root_path):
    for name in files:
        if name.split('.')[0].isdigit():
            tracks_dict[int(name.split('.')[0])] = os.path.join(path, name)
track_id_list = tracks_dict.keys()
track_id_set = set(track_id_list)
```

```python
tracks = utils.load('./data/fma_metadata/tracks.csv')
tracks.head()
```

```python
tracks = tracks['track'][['genre_top', 'title']]
tracks = tracks.loc[track_id_list]
```

```python
tracks = tracks.rename(columns = {'genre_top': 'genre'})
tracks.head()
```

```python
tracks['genre'].value_counts()
```

```python
tracks['genre'] = tracks['genre'].cat.remove_unused_categories()
```
