```python
import pandas as pd
```

```python
tracks = pd.read_csv('./data/final_tracks.csv')
features = pd.read_csv('./data/features.csv')
```

```python
tracks.head()
```

```python
merged_df = tracks.merge(features, how = 'inner', right_on = 'track_id', left_on = 'track_id')
merged_df.head()
```

```python
merged_df.to_csv('./data/merged_data.csv')
```
